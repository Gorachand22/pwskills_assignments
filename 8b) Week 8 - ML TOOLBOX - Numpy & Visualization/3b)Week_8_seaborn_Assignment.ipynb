<p style=" color : #4233FF"><b>Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.</b></p>

Seaborn is a popular data visualization library in Python that provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn along with their uses:

1. Scatter Plot:
   Use: To visualize the relationship between two continuous variables. Each data point is represented as a dot on the plot, and it helps to identify patterns and correlations between the variables.

2. Bar Plot:
   Use: To display the distribution of categorical data. It shows the mean or other summary statistic for different categories and allows for easy comparison between groups.

3. Histogram:
   Use: To visualize the distribution of a single continuous variable. It divides the data into bins and shows the frequency or count of data points in each bin.

4. Box Plot:
   Use: To display the distribution of a continuous variable along with the summary statistics (median, quartiles, and outliers). It is useful for identifying the presence of outliers and visualizing the spread and central tendency of the data.

5. Heatmap:
   Use: To visualize the correlation between multiple variables in a tabular form. It is particularly useful for exploring relationships in large datasets and identifying patterns of high and low values.


<p style=" color : #4233FF"><b>Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.</b></p>

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(8, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Add title and labels
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

<p style=" color : #4233FF"><b>Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.</b>

Note: pclass, age, and fare are columns in the titanic dataset.</p>

```python
# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots
plt.figure(figsize=(12, 6))  

# Box plot for 'age' vs 'pclass'
plt.subplot(1, 2, 1)  # Create the first subplot
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Box plot for 'fare' vs 'pclass'
plt.subplot(1, 2, 2)  # Create the second subplot
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

<p style=" color : #4233FF"><b>Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.</b>
</p>

```python
# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True, bins=30)

# Add title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

<p style=" color : #4233FF"><b>Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.</b>
</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue
sns.pairplot(iris_data, hue="species")

# Show the plot
plt.show()
```

<p style=" color : #4233FF"><b>Que 6: Use the "flights" dataset from seaborn to plot a heatmap.</b>
</p>

```python
# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Create a pivot table using keyword arguments
flights_pivot = flights_data.pivot(index="month", columns="year", values="passengers")

# Create the heatmap
plt.figure(figsize=(10, 7))  
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)

# Add title and labels
plt.title("Heatmap of Flight Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
