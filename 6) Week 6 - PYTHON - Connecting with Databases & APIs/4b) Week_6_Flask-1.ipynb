<p style=" color : #4233FF"><b>Q1. What is Flask Framework? What are the advantages of Flask Framework?<br></b></p>

Flask is a micro web framework for Python, designed to be lightweight and easy to use. It provides the necessary tools and libraries to build web applications and APIs quickly and efficiently. Flask follows the Model-View-Controller (MVC) architectural pattern, but it is flexible enough to adapt to other design patterns as well.

`Advantages of Flask Framework:`

1. Lightweight and Minimalistic: Flask is known for its simplicity and minimalistic approach. It does not impose any specific project structure or dependencies, allowing developers to have more flexibility and control over their applications.

2. Easy to Learn and Use: Flask has a straightforward and intuitive API, making it easy for developers, especially beginners, to get started quickly. The framework provides clear documentation and a small learning curve.

3. Extensibility: Flask is highly extensible, allowing developers to add additional functionalities and libraries as needed. It supports various Flask extensions and third-party libraries, enabling developers to leverage a wide range of tools and features.

4. Flexible Templating: Flask supports Jinja2, a powerful templating engine that allows developers to create dynamic and reusable HTML templates. It provides features like template inheritance, macros, and filters, making it easier to generate dynamic content.

5. Modular Design: Flask follows a modular design, allowing developers to break their applications into smaller, reusable components called blueprints. This modular approach promotes code organization and maintainability.

6. Testing Support: Flask provides built-in support for unit testing, making it easier to write test cases and perform automated testing of web applications. It integrates well with popular testing frameworks and tools.

7. Active Community and Ecosystem: Flask has a vibrant and active community of developers, which means there is ample support, resources, and extensions available. The Flask ecosystem offers numerous plugins, libraries, and tutorials that can help developers accelerate their development process.

8. Scalability: Flask is well-suited for building scalable applications. With the help of Flask extensions and deployment options like Gunicorn and uWSGI, Flask applications can handle high traffic loads and be deployed in production environments.

Overall, Flask's advantages lie in its simplicity, flexibility, extensibility, and ease of use, making it a popular choice for developing web applications and APIs in Python.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

<p style=" color : #4233FF"><b>Q3. What is App routing in Flask? Why do we use app routes?<br></b></p>

n Flask, app routing refers to the process of mapping URL routes to the corresponding functions or view handlers in a Flask application. It allows you to define different routes or URLs that the application can handle and specify the corresponding function or view that should be executed when a particular URL is accessed.

The @app.route() decorator is used in Flask to define routes. It is placed before a function declaration and specifies the URL pattern or route that should trigger the execution of that function.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'This is the contact page.'

if __name__ == '__main__':
    app.run()
```

<p style=" color : #4233FF"><b>Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.<br></b></p>

In Flask, the url_for() function is used for URL building. It is used to dynamically generate URLs for a given function or endpoint in your Flask application.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/contact')
def contact():
    return 'Contact Page'

@app.route('/user/<username>')
def user_profile(username):
    return f'User Profile: {username}'

@app.route('/login')
def login():
    return 'Login Page'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))                   
        print(url_for('about'))                  
        print(url_for('contact'))                
        print(url_for('user_profile', username='john'))  
        print(url_for('login', next='/dashboard'))       
```
