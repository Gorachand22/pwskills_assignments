<p style=" color : #4233FF"><b>Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.</b></p>

Matplotlib is a data visualization library in Python that is widely used for creating static, animated, and interactive visualizations. It provides a comprehensive set of tools for creating various types of plots, charts, and graphs.

Here are five common plots that can be created using the Pyplot module of Matplotlib:

1. `Line plot:` A plot that displays data points connected by straight lines, suitable for showing trends or continuous data.
2. `Bar plot:` A plot that uses rectangular bars to represent data values, suitable for comparing discrete categories or quantities.
3. `Scatter plot:` A plot that uses individual data points to represent the relationship between two continuous variables, suitable for exploring correlations or patterns.
4. `Histogram:` A plot that displays the distribution of a single variable by dividing it into bins and counting the number of occurrences in each bin.
5. `Pie chart:` A circular plot that represents the proportion or percentage of each category in a dataset, suitable for displaying relative proportions.

Scatter plot that uses individual data points to represent the relationship between two continuous variables, suitable for exploring correlations or patterns.

````python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.show()
````

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots in a single figure, making it easier to compare and analyze different data visualizations.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Create a 2x2 grid, select the first subplot
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Select the second subplot
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Select the third subplot
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Select the fourth subplot
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
````

<p style=" color : #4233FF"><b>Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.</b>

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```
</p>

A bar plot, also known as a bar chart, is a graphical representation that uses rectangular bars to compare and display categorical data. Each bar represents a category, and the height or length of the bar corresponds to the value or frequency of that category.

Bar plots are used for several purposes:

1. Visualizing categorical data: Bar plots are particularly useful for displaying and comparing categorical data. They provide a clear and intuitive way to represent the distribution or frequency of different categories.
2. Comparing data across categories: Bar plots allow for easy visual comparison of values or frequencies between different categories. It is effective in identifying patterns, trends, or disparities among the categories.
3. Presenting summary statistics: Bar plots can also be used to represent summary statistics, such as means, medians, or percentages, associated with different categories. This helps in summarizing and presenting key information about the dataset.
4. Making data-driven decisions: By visualizing data using bar plots, it becomes easier to understand and interpret the information, which can aid in making data-driven decisions, identifying areas for improvement, or detecting anomalies.

In a bar plot, the x-axis typically represents the categories, while the y-axis represents the values or frequencies associated with each category. The length or height of the bars represents the values being compared. The bars can be arranged vertically (column/bar chart) or horizontally (barh chart) based on the preference and suitability of the data.

````python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(1,2,1)
plt.bar(company, profit ) # bar plot
plt.xlabel('company')
plt.ylabel('profit')
plt.title('company vs profit')

plt.subplot(1,2,2)
plt.barh(company, profit) # horizontal bar plot
plt.xlabel('company')
plt.ylabel('profit')
plt.title('company vs profit')
plt.tight_layout()
plt.show()
````

```python
box1 = np.random.normal(100, 10, 200)

```
</p>

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's summary statistics. It displays the distribution of a continuous variable through five key summary measures: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. The box plot provides a visual summary of the data's central tendency, variability, and skewness.

The key components of a box plot are as follows:

The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). It contains the middle 50% of the data.
A vertical line inside the box represents the median (Q2), which divides the dataset into two equal halves.
"Whiskers" extend from the box to the minimum and maximum values, excluding any outliers.
Outliers, if present, are represented as individual data points beyond the whiskers.
Box plots are used for several purposes:

Visualizing data distribution: Box plots provide a concise summary of the dataset's spread, skewness, and presence of outliers.
Comparing multiple distributions: Multiple box plots can be plotted side by side to compare different groups or categories within the dataset.
Identifying outliers: Box plots can help identify potential outliers in the data, which are represented as individual points beyond the whiskers.
Assessing symmetry and skewness: By observing the box plot's symmetry or asymmetry, you can infer the skewness of the data distribution.

````python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.grid(True)
plt.show()
````
