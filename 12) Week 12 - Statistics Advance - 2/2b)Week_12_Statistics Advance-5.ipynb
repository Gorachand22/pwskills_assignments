<p style=" color : #4233FF"><b>Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.</b></p>

```python
import scipy.stats as stats

# Given data
sample_mean = 50
sample_std = 5
confidence_level = 0.95
sample_size = 100  # Assuming a sample size of 100 for illustration

# Calculate the t-critical value for the given confidence level and degrees of freedom
degrees_of_freedom = sample_size - 1
t_critical = stats.t.ppf(1 - (confidence_level / 2), df=degrees_of_freedom)

# Calculate the margin of error
margin_of_error = t_critical * (sample_std / (sample_size ** 0.5))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the confidence interval
print("95% Confidence Interval:", confidence_interval)
```

<p style=" color : #4233FF"><b>Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.</b></p>

```python
import scipy.stats as stats

# Observed frequencies of colors in the sample
observed_frequencies = [45, 42, 56, 27, 26, 54]  # Example values, replace with actual data

# Expected distribution percentages
expected_percentages = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]

# Total number of observations
total_observations = sum(observed_frequencies)

# Calculate expected frequencies based on the expected distribution
expected_frequencies = [percentage * total_observations for percentage in expected_percentages]

# Perform the chi-square test
chi_square_stat, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# Significance level
significance_level = 0.05

# Print the results
print("Chi-Square Statistic:", chi_square_stat)
print("p-value:", p_value)

# Compare p-value with significance level
if p_value < significance_level:
    print("Reject the null hypothesis: The distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors matches the expected distribution.")
```

<p style=" color : #4233FF"><b>Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.</b>
</p>

```python
import scipy.stats as stats

# Sample size
n = 500

# Number of individuals who smoke
x = 60

# Calculate the sample proportion
p_hat = x / n

# Calculate the standard error
se = (p_hat * (1 - p_hat) / n) ** 0.5

# Confidence level
confidence_level = 0.95

# Calculate the z-score for the given confidence level
z = stats.norm.ppf(1- ((1 - confidence_level) / 2))

# Calculate the margin of error
margin_of_error = z * se

# Calculate the confidence interval
lower_limit = p_hat - margin_of_error
upper_limit = p_hat + margin_of_error

# Print the results
print("Sample Proportion (p_hat):", p_hat)
print("Standard Error (SE):", se)
print("Z-Score:", z)
print("Margin of Error:", margin_of_error)
print("95% Confidence Interval:", (lower_limit, upper_limit))
```

<p style=" color : #4233FF"><b>Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.</b>
</p>

```python
import scipy.stats as stats

# Sample statistics
sample_mean = 75
sample_std = 12
sample_size = 25

# Confidence level
confidence_level = 0.90

# Calculate the standard error
se = sample_std / (sample_size ** 0.5)

# Calculate the t-score for the given confidence level and degrees of freedom
degrees_of_freedom = sample_size - 1
t = stats.t.ppf(1 - (1 - confidence_level) / 2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error = t * se

# Calculate the confidence interval
lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

# Print the results
print("Sample Mean:", sample_mean)
print("Standard Error (SE):", se)
print("Degrees of Freedom:", degrees_of_freedom)
print("t-Score:", t)
print("Margin of Error:", margin_of_error)
print("90% Confidence Interval:", (lower_limit, upper_limit))
```

<p style=" color : #4233FF"><b>Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.</b>
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Create a range of x values
x = np.linspace(0, 30, 1000)

# Calculate the chi-square probability density function (PDF)
chi2_pdf = stats.chi2.pdf(x, df)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, chi2_pdf, label=f'Chi-Square PDF (df={df})')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
plt.fill_between(x_fill, stats.chi2.pdf(x_fill, df), color='orange', alpha=0.5, label='Chi-Square = 15')

# Add labels and title
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

<p style=" color : #4233FF"><b>Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.</b>
</p>

Given:
- Sample size (n) = 1000
- Number of people who prefer Coke (x) = 520
- Confidence level = 99%

Sample Proportion (p̂) = x / n = 520 / 1000 = 0.52

Critical Value (Z) for a 99% confidence level is approximately 2.576.

Margin of Error (E) = Z x √({p̂(1-p̂)} / n) = 2.576 x √({0.52(1-0.52)}/ 1000 )≈ 0.034

Confidence Interval = (p̂ ± E ) = ( 0.52 ±  0.034 ) ≈ (0.486, 0.554)

`Interpretation:` We are 99% confident that the true proportion of people in the population who prefer Coke is between 0.486 and 0.554.

```python
import numpy as np
import scipy.stats as stats

# Given data
sample_size = 1000
sample_proportion = 520 / sample_size
confidence_level = 0.99

# Calculate the critical value (Z) for the confidence level
critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = critical_value * np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

# Display the results
print(f"Sample Proportion: {sample_proportion}")
print(f"Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

<p style=" color : #4233FF"><b>Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.</b>
</p>

Given data:
- Observed tails (O_tails) = 45
- Total trials (n) = 100
- Significance level (alpha) = 0.05

Expected frequencies for a fair coin (50 tails and 50 heads for a total of 100 trials).

`Step 1: Set Up Hypotheses`
- Null Hypothesis (H_0): The coin is fair (observed frequencies match expected frequencies).
- Alternative Hypothesis (H_1): The coin is biased towards tails (observed frequencies do not match expected frequencies).

`Step 2: Calculate the Expected Frequencies`
- Expected tails (E_tails) = E_heads = n/2 = 100/2 = 50

`Step 3: Calculate the Chi-Square Statistic`<br>
chi^2 = ∑(O - E)^2/E<br>
chi^2 = (45 - 50)^2/50 + (55 - 50)^2/50 = 0.5 + 0.5 = 1 

`Step 4: Degrees of Freedom`<br>
Degrees of freedom (df) = Number of categories - 1 = 2 - 1 = 1

`Step 5: Calculate the Critical Chi-Square Value`<br>
For a significance level of 0.05 and 1 degree of freedom, using a chi-square distribution table or calculator, the critical chi-square value is approximately 3.841.

`Step 6: Compare Chi-Square Statistic with Critical Value`<br>
Since (chi^2 = 1) is less than the critical value of 3.841, we fail to reject the null hypothesis.

In conclusion, based on the manual calculation of the chi-square goodness of fit test, we fail to reject the null hypothesis. This suggests that the observed frequencies of tails and heads in the coin flips do not significantly deviate from the expected frequencies of a fair coin at the 0.05 significance level.

```python
import scipy.stats as stats

# Given data
observed_tails = 45
total_trials = 100
alpha = 0.05

# Expected frequency for tails in a fair coin
expected_tails = total_trials / 2

# Calculate the expected frequency for heads (also total_trials / 2)

# Observed frequencies
observed_frequencies = [observed_tails, total_trials - observed_tails]

# Expected frequencies for a fair coin
expected_frequencies = [expected_tails, total_trials - expected_tails]

# Calculate the chi-square statistic
chi_square_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Calculate the critical chi-square value
degrees_of_freedom = len(observed_frequencies) - 1
critical_chi_square_value = stats.chi2.ppf(1 - alpha, df=degrees_of_freedom)

# Compare the chi-square statistic with the critical chi-square value
if chi_square_statistic > critical_chi_square_value:
    print("Reject the null hypothesis. The coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis. The coin is fair.")
    
print("Chi-Square Statistic:", chi_square_statistic)
print("Critical Chi-Square Value:", critical_chi_square_value)
print("P-Value:", p_value)
```

<p style=" color : #4233FF"><b>Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.</b>
</p>

`Step 1: Set Up Hypotheses`
- Null Hypothesis (H_0): There is no association between smoking status and lung cancer diagnosis.
- Alternative Hypothesis (H_1): There is an association between smoking status and lung cancer diagnosis.

`Step 2: Calculate the Expected Frequencies`
- Total observations (n) = 60 + 140 + 30 + 170 = 400
- Expected frequency for "Smoker" and "Lung Cancer: Yes":
  E{Smoker, Yes} = (Total in Smoker row x Total in Lung Cancer: Yes column) / n = {200 x 90 /400} = 45
- Similarly, calculate expected frequencies for other cells.

`Step 3: Calculate the Chi-Square Statistic`<br>
chi^2 = ∑{(O - E)^2}{E}<br>
chi^2 = {(60 - 45)^2 / 45} + {(140 - 155)^2 / 155} + {(30 - 45)^2}/45 + {(170 - 155)^2}/155 =  10.71

`Step 4: Degrees of Freedom`<br>
Degrees of freedom (df) = (# of rows - 1) \times (# of columns - 1) = 1

`Step 5: Calculate the Critical Chi-Square Value`<br>
For a significance level of 0.05 and 1 degree of freedom, using a chi-square distribution table or calculator, the critical chi-square value is approximately 3.841.

`Step 6: Compare Chi-Square Statistic with Critical Value`<br>
Since (chi^2 = 10.71 ) is greater than the critical value of 3.841, we reject the null hypothesis.

In conclusion, based on the manual calculation of the chi-square test for independence, we reject the null hypothesis and conclude that there is a significant association between smoking status and lung cancer diagnosis at the 0.05 significance level.

```python
import scipy.stats as stats

# Given data
observed = [[60, 140], [30, 170]]
alpha = 0.05

# Perform the chi-square test
chi_square_statistic, p_value, dof, expected = stats.chi2_contingency(observed)

# Calculate the critical chi-square value
critical_chi_square_value = stats.chi2.ppf(1 - alpha, df=dof)

# Compare the chi-square statistic with the critical chi-square value
if chi_square_statistic > critical_chi_square_value:
    print("Reject the null hypothesis. There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between smoking status and lung cancer diagnosis.")
    
print("Chi-Square Statistic:", chi_square_statistic)
print("Critical Chi-Square Value:", critical_chi_square_value)
print("P-Value:", p_value)
```

<p style=" color : #4233FF"><b>Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.</b>
</p>

Certainly, let's calculate the 95% confidence interval for the true proportion of people who are satisfied with their job manually:

Given data:
- Sample proportion (p̂) = 0.65
- Sample size (n) = 500
- Confidence level = 95%
- Critical z-value (z) for a 95% confidence level ≈ 1.96

`Step 1: Calculate the Standard Error`
Standard Error = √(p̂.(1-p̂)/n)
Standard Error = √{{0.65 . (1 - 0.65)}/500} ​≈ 0.0232 

`Step 2: Calculate the Margin of Error`
Margin of Error = z x {Standard Error}
Margin of Error = 1.96 x 0.0232 ​≈ 0.0454

`Step 3: Calculate the Confidence Interval`
Lower Bound: ( p̂ - Margin of Error)
Upper Bound: ( p̂ + Margin of Error)

Lower Bound = 0.65 - 0.0454 ​≈ 0.6046
Upper Bound = 0.65 + 0.0454 ​≈ 0.6954

The 95% confidence interval for the true proportion of people who are satisfied with their job is approximately `(0.6046, 0.6954)`

This means that we are 95% confident that the true proportion of people who are satisfied with their job lies within the interval `(0.6046, 0.6954)` based on the given sample.

```python
import math

# Given data
sample_proportion = 0.65
sample_size = 500
confidence_level = 0.95

# Calculate the critical z-value
critical_z = 1.96

# Calculate the standard error
standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the margin of error
margin_of_error = critical_z * standard_error

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print("95% Confidence Interval:", (lower_bound, upper_bound))
```

<p style=" color : #4233FF"><b>Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.</b>
</p>

Certainly, let's conduct the t-test manually step by step:

Given data:
- Sample size (n) = 30
- Sample mean (bar_x) = 72
- Sample standard deviation (s) = 10
- Population mean hypothesis (mu_0) = 70
- Significance level (alpha) = 0.05

`Step 1: Set Up Hypotheses`
- Null Hypothesis (H_0): The population mean is not significantly different from 70 (mu = \mu_0).
- Alternative Hypothesis (H_1): The population mean is significantly different from 70 (mu not equall to mu_0).

`Step 2: Calculate the Test Statistic`<br>
t = {bar_x - mu_0}/{{s}/{sqrt{n}}}<br>
t = {72 - 70}/{{10}/{sqrt{30}}} = 1.897

`Step 3: Degrees of Freedom`<br>
Degrees of freedom (df) = Sample size - 1 = 30 - 1 = 29

`Step 4: Calculate the Critical t-Value`<br>
For a two-tailed test and a significance level of 0.05, we divide the significance level by 2: alpha = {0.05}/{2} = 0.025.<br>
Using a t-table or calculator, the critical t-value for ( alpha = 0.025 ) and ( df = 29 ) is approximately 2.045.

`Step 5: Compare the Test Statistic with Critical Value`<br>
Since the absolute value of the test statistic (|t| = 1.897) is less than the critical value of 2.045, we fail to reject the null hypothesis.

In conclusion, based on the manual calculation using the t-test, we fail to reject the null hypothesis. This suggests that there is no significant difference between the sample mean and the population mean of 70 at the 0.05 significance level.

```python
import scipy.stats as stats

# Given data
sample_size = 30
sample_mean = 72
sample_std_dev = 10
population_mean_hypothesis = 70
significance_level = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean_hypothesis) / (sample_std_dev / (sample_size ** 0.5))

# Calculate degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical t-value (two-tailed test)
alpha = significance_level / 2
critical_t_value = stats.t.ppf(1 - alpha, df=degrees_of_freedom)

# Compare the t-statistic with the critical t-value
if abs(t_statistic) > critical_t_value:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the population mean.")

print("T-Statistic:", t_statistic)
print("Critical T-Value:", critical_t_value)
```
